```python
import numpy as np
from PIL import Image
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
img_src = Image.open(f'tiles/6/42/42.png')
```

```python
cm = mpl.cm.get_cmap('plasma')
im = np.array(img_src.convert("L"))
im = cm(im)
im = np.uint8(im * 255)
im = Image.fromarray(im)
im
```

```python
im.mode
```

```python
# Now we need to add alpha channel
src = np.array(im)
src = src[:, :, :3]
src
```

```python
alpha = np.array(img_src)
alpha = alpha[:, :, 1:2]
```

```python
res = np.concatenate((src, alpha), axis=2)
res
```

```python
ans = Image.fromarray(res)
ans
```
