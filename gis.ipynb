```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import matplotlib.image as mpimg

Image.MAX_IMAGE_PIXELS = None

img = np.loadtxt("DNI.asc", skiprows=6)
```

```python

def world2Pixel_yours(lng, lat): # x is Longitude and y is Latitude
  ulX = 46.000000000000
  ulY = 40.000000000000
  xDist = 0.002500000000
  yDist = 0.002500000000

  pixel_x = int((lng - ulX) / xDist)
  pixel_y = int((lat - ulY) / yDist)
  return (pixel_x, pixel_y)
```

```python
world2Pixel_yours(76.161780, 44.621473)
```
